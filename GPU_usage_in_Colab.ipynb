```python
import tensorflow as tf
```

```python
device_name = tf.test.gpu_device_name()
if device_name != '/device:GPU:0':
  raise SystemError('GPU device not found')
print('Found GPU at: {}'.format(device_name))
```

```python
def cpu():
  with tf.device('/cpu:0'):
    # your code
cpu()
```

```python
def gpu():
  with tf.device('/device:GPU:0'):
    # your code
gpu()
```

Using GPU is much faster than CPU in big neural network such as image, sound and natural language processing. It may be that small neural network with few calculations is a poor fit for a GPU, which is not seen big different.
